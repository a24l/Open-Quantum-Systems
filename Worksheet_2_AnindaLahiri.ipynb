## 1. 
Hamiltonian dynamics for mixed states (in the Schrödinger picture) is described by the von Neumann equation 

$$\frac{{\rm d}\hat{\rho}}{{\rm d}t} = \frac{1}{{\rm i}\hbar}[\hat{H},\hat{\rho}].$$

Suppose that the Hamiltonian comprises a sum of two terms, $\hat{H} = \hat{H}_0 + \hat{H}_1$. The density matrix in the interaction picture generated by $\hat{H}_0$ is defined by

$$ \hat{\rho}_I(t) = \hat{U}^\dagger_0(t,t_0) \hat{\rho}(t) \hat{U}_0(t,t_0),$$

where $\hat{U}_0(t,t_0) = {\rm e}^{-{\rm i}\hat{H}_0 (t-t_0)/\hbar}$, $\hat{\rho}(t)$ is the density matrix in the Schrödinger picture, and $t_0$ is an (arbitrary) initial time. Show that the state in the interaction picture obeys a differential equation of the form

$$ \frac{{\rm d}\hat{\rho}_I}{{\rm d}t} = \frac{1}{{\rm i}\hbar} [\hat{H}_I(t), \hat{\rho}_I(t)],$$

and give an explicit expression for the interaction-picture Hamiltonian $\hat{H}_I(t)$. What is the general solution of this equation in terms of the initial condition $\hat{\rho}(t_0)$?

## sol: 
consider $$\hat{\rho}_{I}(t) =\hat{U}^{\dagger}_{0}(t,t_{0}) \hat{\rho}_{I}(t)\hat{U}_{0}(t,t_{0})$$
then
$$ \displaystyle \dot{\hat{\rho}_{I}}(t) = e^{i\hat{H_0}(t-t_0)/\hbar}  \dot{\hat{\rho}}(t)  e^{-i\hat{H_0}(t-t_0)/\hbar} + e^{i\hat{H_0}(t-t_0)/\hbar} \left( \frac{-i H_0}{\hbar}\hat{\rho}(t)\right) e^{-i\hat{H_0}(t-t_0)/\hbar} + e^{i\hat{H_0}(t-t_0)/\hbar}  \left( \frac{i H_0}{\hbar}\hat{\rho}(t)\right)   e^{-i\hat{H_0}(t-t_0)/\hbar}  $$ <br>
$$ \displaystyle \implies \dot{\hat{\rho}_{I}}(t) =  \frac{-i}{\hbar} e^{i\hat{H_0}(t-t_0)/\hbar} \left[-H_0 \hat{\rho}(t) + [\hat{H},\hat{\rho}(t)] + H_0 \hat{\rho}(t) \right] e^{-i\hat{H_0}(t-t_0)/\hbar}$$ 
$$ =  \frac{-i}{\hbar} e^{i\hat{H_0}(t-t_0)/\hbar} \left[-H_0 \hat{\rho}(t) + [H_{0}+H_{I},\hat{\rho}(t)] + H_0 \hat{\rho}(t) \right] e^{-i\hat{H_0}(t-t_0)/\hbar} $$ 
$$ =  \frac{-i}{\hbar} e^{i\hat{H_0}(t-t_0)/\hbar} \left[H_{I},\hat{\rho}(t) \right] e^{-i\hat{H_0}(t-t_0)/\hbar} $$
$$ =  \frac{-i}{\hbar} \left( H_{I}e^{i\hat{H_0}(t-t_0)/\hbar} \hat{\rho}(t) e^{-i\hat{H_0}(t-t_0)/\hbar} -  e^{i\hat{H_0}(t-t_0)/\hbar} \hat{\rho}(t) e^{-i\hat{H_0}(t-t_0)/\hbar}H_{I} \right)$$

$$ = \frac{-i}{\hbar}\left[H_{I},\hat{\rho}_{I}(t)\right]$$ <br>

where $H_{I}(t) = e^{i\hat{H_0}(t-t_0)/\hbar} \hat{H_{I}} e^{-i\hat{H_0}(t-t_0)/\hbar}   $

The following function samples a "biased coin", which gives the result 0 (heads) with probability p and gives the result 1 (tails) with probability 1-p. Play around with it and try to understand how it works.

```python
from random import random
def biased_coin(p):
    ''' biased coin function'''
    r = random()
    if r<p:
        return 0
    else:
        return 1
```

(b) Write an analogous function that samples a "three-sided coin", which yields the result 0 with probability $p_0$, 1 with probability $p_1$, and 2 with probability $1-p_0-p_1$. Plot a histogram of 200 samples of the output  for $p_0=0.2$ and $p_1 = 0.5$.

```python
def biased_three_sided_coin(p0,p1):    
    ''' function for three biased sided coin '''
    r = random()
    if r<p0:
        return 0
    elif r<p0+p1:  # taking probability for p0=1 and p1=2
        return 1
    else:
        return 2
    
# use the parameters of two sided coin and select p0 and p1 from above
n = 200
p0 = 0.2
p1 = 0.5

# run the results for 200 iterations using p0 = 0.2 and p1=0.5
results = [biased_three_sided_coin(p0,p1) for i in range(n)]

#plot the results in histogram
plt.figure(figsize=(8,6))
plt.hist(results, bins=9)
plt.grid(axis='y')
plt.xticks(range(3))
plt.xlabel('Result', fontsize= 15)
plt.ylabel('Frequency', fontsize= 15)
plt.title(f'Three sided biased coin for {n} shots (p0 = {p0}, p1 = {p1})', fontsize= 15)
plt.show()
```

(a) Consider the evolution starting from the initial state $\hat{\rho}(0) = |+\rangle\langle +|$. Plot the expectation value $\langle \hat{\sigma}_x\rangle$ as a function of time for each realisation (put them all on one plot). 

```python
#input the given parameters
w = 1
g = 0.1*w 
# since omega gives a gaussian distribution choose omega value as a normal distrubution
W = np.random.normal(w, g, 100) # random number array of 100

#choose the time interval
min_w0t=0
max_w0t = 30
interval = 0.01
w0t = np.arange(min_w0t,max_w0t,interval) # make the x-axis 

#define all the ncecessary functions
def time_evolution_operator(matrix,time_interval,const=1):
    '''return the time evolution operator for a given time interval and matrix'''
    return(expm(-1j*const*time_interval*matrix))
        
def hamilton_matrix(u=[0,0,1],hw=0.2):
    '''computes the hamiltonian matrix (in this case only for sigma_z)'''
    sigma_x = u[0]*X
    sigma_y = u[1]*Y
    sigma_z = u[2]*Z
    return (hw/2)*(sigma_x+sigma_y+sigma_z)

states = []   # create an empty list
states.append(plus)  #store the |+> state

# define function to extract rxpectation value
def expectation_value(operator,state):
    ''' returns the expectation value for a given state and operator'''
    result = operator.T.conj() @ state @ operator
    return float(result[0,0].real)

def expectation_value_temporal(max_time, states, state,time_operator,time_interval = 0.01):
    ''' returns the expectation value per discrete time window for a given state and operator'''
    max_bucket = int(max_time/time_interval)
    # set up my empty results table 
    results = np.zeros((max_bucket,len(states)))
    # enumerate  states into a dict to update in a loop
    init_state_dict = dict(enumerate(states))
    # loop over each time window and initial state
    for i in range(max_bucket):
        for j in range(len(init_state_dict)):
            results[i,j] = expectation_value(init_state_dict[j],state.real)
            init_state_dict[j] = time_operator @ init_state_dict[j]
    # return the output
    return results

def gaussian_expectation_temporal(observable):
    '''run the simulation for a given observable'''
    w = np.random.choice(W,1)   #choose random omega values from the array of 100
    w_list.append(w)  # put it in the list 
    U_t = time_evolution_operator(hamilton_matrix(hw = w),0.01)
    return expectation_value_temporal(max_w0t, states, observable,U_t)

def plot_gaussian_expectation(n,results,observable):
    '''plot expectation value for the n simulations'''
    plt.figure(figsize = (10.0,7.0))
    for i in range(n):
        plt.plot(w0t,results[i],label = '$\\omega$ = '+str(w_list[i]))
    plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.4),fontsize=16)
    labels = [r'$'+str(i)+'\pi$' for i in range(1,11)]
    labels.insert(0,'$0$') 
    plt.xlabel('$\omega_0$t', fontsize=15)
    plt.grid(True)
    plt.xlim([0,30])
    plt.ylabel('$\langle \hat{\sigma}_x\\rangle$', fontsize=15)
    plt.title( f'Expectation value of $\sigma_x$ for {n} simulations', fontsize=15)
    plt.show()
    

# define my empty w list    
w_list = []
# define number of simulations
n=20

#run the simulations
results = [gaussian_expectation_temporal(X) for i in range(n)]

# plot the results
plot_gaussian_expectation(n,results,X) 
```
